```python
import pandas as pd
from matplotlib import pyplot as plt
df = pd.read_csv('/content/company_sales_data.csv')
df.head()
x = df['month_number'].tolist()
y = df['total_profit'].tolist()
plt.plot(x,y,label = "total monthly profit in a year")
plt.xticks(x)
plt.xlabel("month number")
plt.ylabel("total profit")
#rint(df)
```
